```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv('ff_race_50.csv',sep=";")
```

```python
df
```

```python
df.head()
```

```python
df2 = df.dropna(axis=1)
```

```python
df2.head()
```

```python
df2.loc[:, 'fullname'] = df2['First'] + ' ' + df2['Last']
```

```python
df2.head()
```

```python
df2.loc[:,'Time'] = pd.to_timedelta(df2['Time'])
```

```python
df2.loc[:,'Total_Minutes'] = df2['Time'].dt.total_seconds() / 60
```

```python
df2.loc[:,'Total_Minutes'] = df2['Total_Minutes'].round().astype(int)
```

```python
df2.head()
```

```python

def get_lat_long(city, state):
  address = f"{city}, {state}"
  try:
    geolocator = Nominatim(user_agent="running", timeout=10)
    location = geolocator.geocode(address)
    if location:
      return location.latitude, location.longitude
    else:
      return None, None
  except GeocoderTimedOut:
    return None, None
```

```python
df2['latitude'],df2['longitude'] = zip(*df2.apply(lambda x: get_lat_long(x['City'], x['state']) , axis=1))
```

```python
df2.head()
```

```python
df2['latlong'] = df2['latitude'].astype(str) +', '+ df2['longitude'].astype(str)
```

```python
df2.head()
```

```python
df2.to_csv('cleandata2_output.csv', index=False)
```
